```python
import pandas as pd
import numpy  as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("canada_per.csv")
df
```

```python
df = df.rename(columns={'per capita income (US$)': 'per_capita_income'})
df
```

```python
%matplotlib inline
plt.xlabel('year')
plt.xlabel('per_capita_income')
plt.scatter(df.year, df.per_capita_income, color='red', marker ='+')
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income) # year ko value layera per capita predict garxa
```

```python
reg.predict(pd.DataFrame([[2020]], columns=['year']))    # formula for linear model: y = B0 + B1*x
```

```python
reg.coef_[0]       #slope calculation m or B1
```

```python
reg.intercept_     #calculates B0 also known as y-intercept
```

```python
%matplotlib inline
plt.xlabel('year',fontsize=20)
plt.ylabel('per_capita_income',fontsize=20)
plt.scatter(df.year,df.per_capita_income, color ='green', marker ='*')
plt.plot(df.year,reg.predict(df[['year']], color = 'blue')
```
