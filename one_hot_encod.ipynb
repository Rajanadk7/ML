```python
import pandas as pd
```

```python
df = pd.read_csv("homeprice.csv")
df
```

Using pandas to create dummy variables

```python
dummies = pd.get_dummies(df.town)  # creating a columes for the dummy variables for the catagrorail variables town
dummies
```

```python
bool_cols = dummies.select_dtypes(include='bool').columns   #converting
dummies[bool_cols] = dummies[bool_cols].astype(int)
dummies
```

```python
merged = pd.concat([df, dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns') # droping the column town to reduce categorial variable and column west windsor to reduce multicolinearity
final 
```


Dummy Variable Trap
When you can derive one variable from other variables, they are known to be multi-colinear. Here if you know values of california and georgia then you can easily infer value of new jersey state, i.e. california=0 and georgia=0. There for these state variables are called to be multi-colinear. In this situation linear regression won't work as expected. Hence you need to drop one column.

NOTE: sklearn library takes care of dummy variable trap hence even if you don't drop one of the state columns it is going to work, however we should make a habit of taking care of dummy variable trap ourselves just in case library that you are using is not handling this for you

```python
from sklearn.linear_model import LinearRegression
```

```python
x =final.drop('price', axis ='columns')
x
```

```python
y = final.price
y
```

```python
model = LinearRegression() # 
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
model.score(x,y)  #shows how accurate is our prediction
```


Using sklearn OneHotEncoder
First step is to use label encoder to convert town names into numbers

```python
df
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
dfle.town = le.fit_transform(dfle.town) # fits numerical value to the town column label encoder and fir_transdorm function
dfle
```

```python
x= dfle[['town','area']].values
x
```

```python
y = dfle.price.values
y
```

Now use one hot encoder to create dummy variables for each of the town

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('town', OneHotEncoder(), [0])], remainder = 'passthrough')
```

```python
x = ct.fit_transform(x)
x
```

```python
x = x[:,1:]
x# jhis means droping 0th column and taking columns others 
```

```python
model.fit(x,y)
```

```python
model.predict([[0,1,3400]])
```

```python
model.predict([[1,0,2800]])
```
